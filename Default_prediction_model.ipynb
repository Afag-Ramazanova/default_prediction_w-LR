### Preprocessing

````python
df['CUSTOMER_CITY'].unique()
# Displays unique values of a column.
# Here we can see there is multiple versions of some City names, we need to equalize the same city formats
````

````python
type(df['CUSTOMER_CITY'][0])
````

````python
df['CUSTOMER_CITY'] = df['CUSTOMER_CITY'].str.replace(',', '')
df['CUSTOMER_CITY'] = df['CUSTOMER_CITY'].str.replace('-', '')
df['CUSTOMER_CITY'] = df['CUSTOMER_CITY'].str.replace('Baku', 'Bakı')
df['CUSTOMER_CITY'] = df['CUSTOMER_CITY'].str.replace('Ismaıly', 'İsmayıllı')
````

````python
df['CUSTOMER_CITY'] = df['CUSTOMER_CITY'].str.upper()
````

````python
df['LOAN_DURATION'] = round(pd.to_numeric((df['DATE_OF_LOAN_MATURITY'] - df['START_DATE']) / np.timedelta64(1, 'M')))
df['Diff'] =  df['TENURE'] - df['LOAN_DURATION']
df['PROLONGATION'] = (df['Diff'] < 0).astype(int)
df = df.drop(['Diff', 'CONTRACT_NO','START_DATE', 'DATE_OF_LOAN_MATURITY'],axis=1)
#here new variable constracted by subtracting date of loan maturity and start date
# then if there is negative difference between Tenure and loan duration it means loan duration
#is prolonged and it most of the time leads to a default
````

````python
df.columns
````

````python
df['CUSTOMER_CITY']= df.groupby('REGION_LOCATION')['CUSTOMER_CITY'].transform(lambda x: x.fillna(x.mode()[0]))
#here filled missing values of city by their corresponding region location's mode value
````

````python
df['LOAN_LOCATION_GROUP'] =(np.where((df['LOAN_LOCATION_GROUP'] == 'null (missing)') ,df['REGION_LOCATION'], df['REGION_LOCATION']))
#here there is one category null missing we replaced it with corresponding region location value
````

````python
import math


# data - data df
# fl - vars list to bin7


y = data['BAD_FLAG'].values.astype('float') # TARGET VARIABLE

data_bin_optb = pd.DataFrame() # BINNED COLUMNS WILL BE HERE
optb_bin_iv_js = pd.DataFrame(columns = ['var', 'iv', 'js'], dtype = 'float') # df information values & jensen-shennon divergence


data_bin_optb['y'] = y

d_splits = dict()
d_woe = dict()


for i in dh_numer_vl: #


    print('\n', "var: ", i)

    x = data[i].values.astype('float')


    optb = OptimalBinning(dtype="numerical", solver="cp"
                          , divergence = 'iv' # binning criteria
                          , max_n_bins= 5

                          , min_bin_size = 0.05 #

                          , min_bin_n_nonevent = 1
                          , min_bin_n_event = 1 #

                          , monotonic_trend = 'auto_asc_desc' # monoton binning
                         )
    optb.fit(x, y)


    binning_table = optb.binning_table
    type(binning_table)
    bin_tbl = pd.DataFrame(binning_table.build())

    binning_table.plot(metric="woe")

    print('\n', "BINNING TABLE: ", '\n', bin_tbl)

    print('\n', 'IV: ', float(bin_tbl[-1:]['IV']), 'JS: ', float(bin_tbl[-1:]['JS']))

    optb_bin_iv_js = optb_bin_iv_js.append({'var': i, 'iv': float(bin_tbl[-1:]['IV']), 'js': float(bin_tbl[-1:]['JS'])}, ignore_index=True)

    arr = optb.splits
    arr = [-math.inf, *arr]
    arr = [*arr, math.inf]
    print('\n', "SPLITS: ", arr) #

    d_splits[i] = arr


    woe_list = bin_tbl['WoE'].head(len(arr)-1)
    woe_list = np.concatenate((woe_list, [float(bin_tbl[-2:-1]['WoE'])])) # woe for null values if any

    d_woe[i] = woe_list


    print('\n', "WOE", '\n', woe_list)

    binned_column = []
    binned_column = pd.cut(x, bins = arr, labels = bin_tbl['WoE'].head(len(arr)-1))
    binned_column = binned_column.add_categories(float(bin_tbl[-2:-1]['WoE'])).fillna(float(bin_tbl[-2:-1]['WoE'])) #binning nulls
    data_bin_optb[i] = binned_column


    print('\n', '-----------------------------------------', '\n')

bin_arrays = pd.DataFrame.from_dict(d_splits, orient='index').transpose()
woe_arrays = pd.DataFrame.from_dict(d_woe, orient='index').transpose()
````
